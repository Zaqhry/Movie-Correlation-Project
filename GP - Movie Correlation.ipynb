# Movie Correlation 

In this Project, we are looking for the different categories that have the highest correlation in relation to the Gross Earnings of movies. 

### Import Libraries & Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline

matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

movie = pd.read_csv('movies.csv')
```

### Look at the data

```python
movie
```

### Check for missing data

```python
# Find number of missing values for each column 

movie.isnull().sum() 
```

```python
# find percentage of missing data 

for col in movie.columns:
    pct_missing = np.mean(movie[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(movie.dtypes)
```

```python
# Are there any Outliers?

movie.boxplot(column=['gross'],color='blue')
```

```python
# Drop duplicates

movie.drop_duplicates()
```

```python
# Order our Data

movie.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Correlation of gross in relation to budget 

sns.regplot(x="budget", y="gross", data=movie, scatter_kws={'color':'mediumpurple'},line_kws={'color':'black'})
```

```python
# Correlation of gross in realtion to score 

sns.regplot(x="score", y="gross", data=movie,scatter_kws={'color':'mediumpurple'},line_kws={'color':'black'})
```

```python
correlation_matrix = movie.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

### Correlation Matrix between all numeric columns


```python
# Pearson's Method

movie.corr(method ='pearson')
```

```python
correlation_matrix = movie.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Kendall's Method 

movie.corr(method ='kendall')
```

```python
correlation_matrix = movie.apply(lambda x: x.factorize()[0]).corr(method='kendall')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Spearman's Method

movie.corr(method ='spearman')
```

```python
correlation_matrix = movie.apply(lambda x: x.factorize()[0]).corr(method='spearman')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

movie.apply(lambda x: x.factorize()[0]).corr(method='pearson')
```

```python
correlation_mat = movie.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
movie_numerized = movie


for col_name in movie_numerized.columns:
    if(movie_numerized[col_name].dtype == 'object'):
        movie_numerized[col_name]= movie_numerized[col_name].astype('category')
        movie_numerized[col_name] = movie_numerized[col_name].cat.codes
        
movie_numerized
```

```python
movie_numerized.corr(method='pearson')
```

```python
correlation_matrix = movie_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```
